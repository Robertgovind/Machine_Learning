```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan_data.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python

#fills 100 at missing place and its wrong way
dataset.fillna(100)
```

```python
dataset.info()
```

```python
# bfill for backward fill and ffill for forward fill
dataset.fillna(method="bfill")
```

```python
#axis =1 for column and axis = 0 for row wise fill
dataset.fillna(method="bfill",axis=1)
```

```python
dataset["dti"].mode()
```

```python
dataset["dti"].fillna(dataset["dti"].mode()[0])
```

```python
dataset.select_dtypes(include="object")
```

```python
#gives all names of columns of object types
dataset.select_dtypes(include="object").columns
```

```python
for i in dataset.select_dtypes(include="object").columns:
    print(i)
```

```python
#fill all nan data of type objects
for i in dataset.select_dtypes(include="object").columns:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

```python
#now there is no missing values
dataset.isnull().sum()
```
