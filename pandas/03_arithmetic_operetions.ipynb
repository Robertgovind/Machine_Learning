```python
import pandas as pd
```

```python
value = pd.DataFrame({"A":[2,3,5,7],"B":[1,2,3,4]})
value
```

```python
value["C"] = value["A"] + value["B"]
value
```

```python
value["D"] = value["A"] - value["B"]
value
```

```python
value["E"] = value["A"] * value["B"]
value
```

```python
value["F"] = value["A"] / value["B"]
value
```
