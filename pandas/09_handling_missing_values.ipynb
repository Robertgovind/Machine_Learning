```python
# Replace and Interpolate functions in Pnadas
import pandas as pd
```

```python
dataset = pd.read_csv("health_rankings.csv")
dataset.head(5)
```

```python
dataset.replace(to_replace=43.0,value=50)
```

```python
dataset.replace(to_replace="US",value="Nepal")
```

```python
dataset.replace("[A-Za-z]","Python",regex=True)
```

```python
dataset.replace(43.0,method="ffill")
```

```python
# Make changes in the original data
dataset.replace(43.0,method="bfill",inplace=True)
dataset
```

```python
# Interpolate
# works with numbers
dataset.interpolate()
```

```python
dataset.interpolate(method="linear")
```

```python
dataset.interpolate(method="linear",axis=0)
# Interpolating column wise
```

```python
dataset.interpolate(limit=2)
```

```python
dataset.interpolate(limit=1000,limit_direction="forward")
```
